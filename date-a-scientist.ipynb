## Data Preparation

### Missing Data
The most essential step to start with the prep is to discard the null values which can introduce noise to our model.

```python
profiles.isnull().sum()
```

### Preprocessing
We can work with the whole dataset, however, this may be inefficient for the later steps, therefore we need to select only variables that we think are relevant to predict our target variable.

```python
cols = ['body_type', 'diet', 'orientation', 'pets', 'religionClean', 'sex', 'job', 'signsClean']
df = profiles[cols].dropna()
df.shape
```

### Dummy Variables
We need to encode our categorical variables. The data now has 81 columns to predict signs.

```python
for col in cols[:-1]:
    df = pd.get_dummies(df, columns=[col], prefix = [col])
```

```python
df.head()
```

```python
df.signsClean.value_counts()
```

We can see that there are no considerable data label imbalances.

### Splitting Data
We will use a 75-25 split for our model.

```python
col_length = len(df.columns)
X = df.iloc[:, 1:col_length] # We select columns 1 - len(col)
Y = df.iloc[:, 0:1]# We select just the first column

val_size = 0.25

from sklearn.model_selection import train_test_split
X_train, X_val, Y_train, Y_val = train_test_split(X,Y, test_size = val_size, random_state= 0)

Y_train = Y_train.to_numpy().ravel()
Y_val = Y_val.to_numpy().ravel()
# we use .ravel() to convert Y_train and Y_val to 1D arrays since many ML models expect that structure for the target variable
```

## Pediction
### Model building
We will create three multi-class models to make predictions with our data: Logistic Regression, Decision Trees, KNN.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

## Evaluation Metrics
We will be using different evaluation metrics for our models such as:
- **Accuracy** : correct values divided by the total values.
- **Recall** : true positives divided by the positive guesses.
- **Precision** : true positives divided by the actual positive values.
- **F1 Score** : a blended score of precision and recall which balances both values.

### Logistic Regression
We begin with the regression models, which in this case is a logistic regression because we are looking to classify our outputs.

```python
lr_model = LogisticRegression(multi_class="multinomial").fit(X_train, Y_train)
lr_predictions = lr_model.predict(X_train)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_train, lr_predictions))
```

This model performs quite poorly considering its accuracy score of 12%, which in this case is near to random guess (1/12 = ~8%).

### K Nearest Neighbor

```python
knn_model = KNeighborsClassifier(n_neighbors = 5).fit(X_train, Y_train)
knn_predictions = knn_model.predict(X_train)
```

```python
print(classification_report(Y_train, knn_predictions))
```

We can see there is a significant improvement from the logistic regression model. The accuracy score for this model is 33%.
